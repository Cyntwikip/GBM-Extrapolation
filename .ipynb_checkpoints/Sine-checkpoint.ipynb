# Sine Wave Extrapolation using GBM

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# np.arange(1,10,100)
X = np.linspace(0, 19.9, 200)
y = np.sin(X*np.pi)

X1 = X[:100]
y1 = y[:100]
X2 = X[99:]
y2 = y[99:]

plt.plot(X1,y1)
plt.plot(X2,y2)
```

```python
from sklearn.ensemble import GradientBoostingRegressor

# generate training matrix
def gen_matrix(x, data_num, window_size):
    output = []
    for i in range(data_num):
        row = np.append(x[i:window_size+i], x[window_size+i])
        output.append(row)
    return pd.DataFrame(output)

window_size = 2
X_train = gen_matrix(y1, 90, window_size)

# train model
model = GradientBoostingRegressor(max_depth=4, learning_rate=0.1, random_state=0)
model.fit(X_train.iloc[:,:-1],X_train.iloc[:,-1])

y_new = y1

num_future = 200
for i in range(num_future):
    y_pred = model.predict([y_new[-window_size:]])
    y_new = np.append(y_new, y_pred)
```

```python
plt.figure(dpi=150)
plt.plot(range(len(y1)), y1)
plt.plot(np.arange(len(y1)-1, len(y1)-1+len(y2)), y2, 'r')
plt.plot(np.arange(len(y1)-1, len(y_new)), y_new[len(y1)-1:len(y_new)], 'co', alpha=0.5)
```

```python
from sklearn.metrics import r2_score

print("R2 score = %f"%r2_score(y2, y_new[99:200]))
```
